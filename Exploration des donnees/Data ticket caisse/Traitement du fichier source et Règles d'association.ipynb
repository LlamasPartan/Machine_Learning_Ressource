## Importation des librairies

```python
import pandas as pd
import numpy as np
```

## Lecture et traitement du DataSet

```python
data = pd.read_csv("Recipe.csv", header=None)
```

```python
data
```

```python
dodo = data.iloc[:, [0,1,2,3,4,5,6,7]]
```

```python
dodo = dodo.dropna()
```

```python
dodo = dodo.drop(0,axis=1)
```

```python
ListeIndex = []
ListeGen = []
for i in dodo.index:
    for a in dodo.columns:
        ListeIndex+=[i]
        ListeGen+=[dodo[a][i]]
```

```python
dodo = pd.DataFrame(data = {"Produit" : ListeGen, "NumTicket" : ListeIndex})
```

## Construction des règles d'association

```python
#tableau croisé 0/1
DataFrame = pd.crosstab(dodo.NumTicket,dodo.Produit)
```

```python
DataFrame
```

```python
#importation de la fonction apriori
from mlxtend.frequent_patterns import apriori
 
#itemsets frequents
freq_itemsets = apriori(DataFrame,min_support=0.025,max_len=3, use_colnames=True)
```
